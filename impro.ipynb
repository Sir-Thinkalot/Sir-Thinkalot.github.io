```python
import imageio
import numpy as np
import scipy.ndimage as ndi
import matplotlib.pyplot as plt

im = imageio.imread('/content/sample_data/hand.png')
weights = np.array(([0.11,0.11,0.11],[0.11,0.11,0.11],[0.11,0.11,0.11]))
weights
```

```python
im_filt = ndi.convolve(im,weights)
fig, axes = plt.subplots(2,1, figsize=(20,10))
axes[0].imshow(im,cmap='gray')
axes[1].imshow(im_filt,cmap='gray')
plt.show()
```

```python
im_Gauss1 = ndi.gaussian_filter(im, sigma=2)
im_Gauss2 = ndi.gaussian_filter(im, sigma=10)

fig, axes = plt.subplots(2, 1, figsize=(20,10))
axes[0].imshow(im_Gauss1,cmap='gray')
axes[1].imshow(im_Gauss2,cmap='gray')
plt.show()
```
